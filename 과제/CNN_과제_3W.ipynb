```python
import cv2
from google.colab.patches import cv2_imshow # cv2.imshow(img)가 안되므로 
from PIL import Image as PILImage
import matplotlib.pyplot as plt

import glob

fig = plt.figure(figsize=(10,10))

dolphin_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/original/dolphin/*')
shark_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/original/shark/*')
whale_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/original/whale/*')
```

```python
# 1번 과제
def imgreading(file_path):
  imgs = cv2.imread(file_path)
  return cv2.cvtColor(imgs, cv2.COLOR_BGR2RGB)

import random
imgs = []
imgnum = range(0, 300)

for i in random.sample(imgnum, 3):
  imgs.append(imgreading(dolphin_img_list[i]))
  imgs.append(imgreading(shark_img_list[i]))
  imgs.append(imgreading(whale_img_list[i]))

rows = 3
columns = 3

fig, axes = plt.subplots(rows, columns, figsize = (rows*3, columns*3))

for num in range(1, rows*columns+1):
  fig.add_subplot(rows, columns, num)
  idx = num - 1

  plt.imshow(imgs[idx], aspect = 'auto')
  plt.xlabel(f'{imgs[idx].shape}', fontsize = 12)

fig.tight_layout()

colors = ['dolphin', 'shark', 'whale']

for folder_idx, ax in enumerate(axes[0]):
  ax.set_title(colors[folder_idx])

for idx, ax in enumerate(axes.flat):
  ax.set_xticks([])
  ax.set_yticks([])
```
